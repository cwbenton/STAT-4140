## Naive Bayes Classifier

Fast and simple classification algorithms that can be used on high-dimensional data. It is built on Bayesian classification methods. Find the probability of a label given some observed features.\
`P(L | Features) = [P(Features | L)P(L)] / P(Features)`\
\
We assume that the features are independent of one another.\
Data from each label is assumed to follow a Gaussian (normal) distribution.

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme()
```

```python
from sklearn.datasets import make_blobs # Helps with making scatterplots.
# Want 100 records, 2 features (2 input variables), centers=2 means I want 2 classes, random_state=2
# makes the data return the same data each time.
X, y = make_blobs(100, 2, centers=2, random_state=2) 
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y)
```

```python
# Randomly make test data since we didn't make any.
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)
ynew = model.predict(Xnew)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
# alpha=0.1 changes the transparency.
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.1)
plt.axis(lim);
```
