```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Accessing features (X) and target (y)
X_iris = iris.data
y_iris = iris.target
```

### 1. Choose the Model Class

```python
# Principal Concept Analysis
from sklearn.decomposition import PCA
```

### 2. Instantiate the Model with Hyperparameters

```python
# In the original dataset, there are 4 x-values. We want to reduce the dimensionality to 2.
model = PCA(n_components=2)
```

### 3. Fit to Data

```python
# Notice y is not specified here.
model.fit(X_iris)
```

### 4. Transform the Data to Two Dimensions

```python
X_2D = model.transform(X_iris)
X_2D.shape
```

```python
# It's difficult to see patterns above 4 dimensions, so it can also help visualize.
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x=X_2D[:,0],y=X_2D[:,1],hue=y_iris)
plt.xlabel("pca1")
plt.ylabel("pca2")
plt.show()
```
