```python
# Put datasets into six different dataframes.
df1 = pd.read_csv("/Users/cwbenton/MATH4140/HW/HW2/datasets/genome-scores.csv")
df2 = pd.read_csv("/Users/cwbenton/MATH4140/HW/HW2/datasets/genome-tags.csv")
df3 = pd.read_csv("/Users/cwbenton/MATH4140/HW/HW2/datasets/links.csv")
df4 = pd.read_csv("/Users/cwbenton/MATH4140/HW/HW2/datasets/movies.csv")
df5 = pd.read_csv("/Users/cwbenton/MATH4140/HW/HW2/datasets/ratings.csv")
df6 = pd.read_csv("/Users/cwbenton/MATH4140/HW/HW2/datasets/tags.csv")
```

```python
'''
We will skim the datasets and look for any miscodings or missing data.
We will most likely have to address rows with NaN values. 
Using df.columns, there are no NaN columns for any df.
Using df.isnull(), we see df3 and df6 have NaN values.
df3 has 219 rows with null tmdbId and df6 has 16 rows with a null tag. We will remove these rows.
'''
print(df3.isnull().sum())
df3 = df3.dropna(axis=0)
print(df3.isnull().sum())
```

```python
'''
We will merge genome-tags (df2) and genome-scores (df1), ensuring all rows of genome-scores are
present. Since I want to ensure all rows of df1 are present, we will do a left join.
'''
genome_merge = pd.merge(df1, df2, on="tagId", how="left")
genome_merge
```

## PART F

```python
# View the contents of our merge in part d
movieId_merge.head()
```

```python
# View the contents of our merge in part e
genome_merge.head()
```

```python
# We will now merge movieId_merge and genome_merge, where we only have the rows with common movieId.
# Since we want the rows with common movieId, we do an inner merge.
main = pd.merge(movieId_merge, genome_merge, on='movieId', how='inner')
main
```

## PART G

```python
# Makes it so the output is not in scientific notation (9 decimals specifically)
pd.set_option('display.float_format', lambda x: '%.5f' % x)
# View the 25th percentile
main["relevance"].describe()
```

```python
# Filter all movies with relevance in the 25th percentile.
# Any value from the min to 0.02275 is in the 25th percentile.
main.loc[main["relevance"] <= 0.02275, "relevance"] = 1.0
main
```

## PART H

```python
# Extract the rightmost digit of tmdbId and put it into new tmdbId_recoded column.
# We need to truncate tmdbId and remove the decimals (we can do this since an ID has no decimals)
main["tmdbId"] = main["tmdbId"].astype(int)

# Now we take our column, treat it as a string, take the last value (str[-1]), then put it as an int.
main["tmdbId_recoded"] = main["tmdbId"].astype(str).str[-1].astype(int)
main
```

```python
# Write this output to a csv file.
main.to_csv("/Users/cwbenton/MATH4140/HW/HW2/cbento23_1h.csv")
```

## PART I

```python
# Find the average relevance for each movie.
# We group by title and find the average relevance. reset_index() makes it a dataframe.
avg_relevance_table = main.groupby("title")["relevance"].mean().reset_index()
avg_relevance_table.columns = ["title", "average relevance"]
avg_relevance_table
```
