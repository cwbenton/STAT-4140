### Assembling

###### Merging
Similar to the join operation in SQL <br/>
Used to get data from different sources and combine them

```python
df_freq1 = pd.read_csv("datasets/yo_frequency_list.csv")
df_freq2 = pd.read_csv("datasets/st_frequency_list.csv")
df_freq1
```

```python
df_freq2
```

```python
pd.merge(df_freq1,df_freq2)
# merge() gives us the rows that are common across both the dataframes.
```

```python
pd.merge(df_freq1,df_freq2,on="word")
```

```python
# Inner joins consider A intersect B. (Think of venn diagrams in set theory)
pd.merge(df_freq1,df_freq2,on="word",how="inner")
```

```python
# Outer joins consider A union B.
pd.merge(df_freq1,df_freq2,on="word",how="outer")
```

```python
# Left Outer join considers A
pd.merge(df_freq1,df_freq2,on="word",how="left")
```

```python
# Right outer join considers B.
pd.merge(df_freq1,df_freq2,on="word",how="right")
```

```python
# What if the column names are different?
df_freq1.columns = ['frequency','words']
```

```python
# If the column names are different in the two tables, we can do this.
pd.merge(df_freq1,df_freq2,left_on="words",right_on="word",how="inner")
```

###### Concatenating
Attaches dataframes vertically/horizontally

```python
# Adding columns to it
pd.concat([df_freq1,df_freq2],axis=1)
```

```python
# Adding rows to it
pd.concat([df_freq1,df_freq2],axis=0)
```

### Data Transformation

#### Handle missing data
Note that missing data conventions change from dataset to dataset. <br/>
In some datasets, you will see a fixed value indicating the missing data value. In others, it can be NaN. <br/>
It is important to read the variable descriptions. 

```python
df_new= pd.read_csv("datasets/air-quality-monitoring-sites-summary.csv")
df_new=df_new.drop(df_new.columns[0],axis=1)
df_new
```

```python
# Removes the rows with all NaN values.
df_new.dropna(how="all")
```

```python
# Replaces the dataframe with the removed NaN rows.
df_new=df_new.dropna(how="all")
```

```python
# Deletes columns that have all NaN values. Default is axis=0
df_new.dropna(axis=1,how='all')
```

```python
df_new= df_new.dropna(axis=1,how='all') #for columns
df_new= df_new.dropna(axis=0,how='all') #for rows
```

```python
df_new
```

```python
# Fill all null values with 0 - Data Imputation
# In this case, this is bad.
df_new.fillna(0)
```

```python
df_new.fillna(method='ffill',axis=0)
#makes more sense than the previous one
#useful for time-series data where the previous records are related to the next ones in time
```

```python
df_new['Altitude (ahd)'].fillna(df_new['Altitude (ahd)'].median())
#useful for numerical columns
#median is more robust to outliers than a mean is
#Similarly a mode can be used when you are using categorical data
```
