### Discretization and Binning <br/>
helps with simplifying data, help understand relationships in data much better and interpretable

```python
df
```

```python
df['Unit price'].describe()
```

```python
bins = [0,10,20,30,40,50]
pd.cut(df['Unit price'], bins)
```

```python
df['price_disc'] = pd.cut(df['Unit price'], bins) #Create new column for bins
df['price_disc'].value_counts() #gives a frequency count of each category
```

```python
#you can also rename these values
bin_names = ['very low', 'low', 'medium', 'high', 'very high']
df['cat_price_disc'] = pd.cut(df['Unit price'], bins, labels=bin_names)
df['cat_price_disc']
```

```python
#you can also split into 'n' different bins
#notice how the bins are split in this case
pd.cut(df['Unit price'], 5)
```

```python
pd.cut(df['Unit price'],5).value_counts()
```

```python
# Quantile Cut
# There cannot be duplicates with qcut. Look at describe(), 10 repeats.
pd.qcut(df['Unit price'], 4)
```

```python
# We can drop duplicates when doing a quantile cut
pd.qcut(df['Unit price'], 4, duplicates='drop')
```

```python
df['Total_q'] = pd.qcut(df['Total'],4)
df['Total_q']
```

```python
mean = df_new['Quantity'].mean()
std = df_new['Quantity'].std()
outliers = df_new[(df_new['Quantity'] - mean).abs() > 3 * std]
```

```python
outliers
```

```python
#how to repeat it for all the columns
for col in df_new:
    mean = df_new[col].mean()
    std = df_new[col].std()
    outliers = df_new[(df_new[col] - mean).abs() > 3 * std]
    print(outliers)
```

### Data Aggregation
Grouping records based on certain conditions and applying various operations on them.<br/>
SPLIT-APPLY-COMBINE Mechanism




<img src="Screenshot.png" alt="Alternative text"/>

```python
df
```

```python
df.groupby('Gender')
```

```python
#Group iteration
for name, group in df.groupby('Gender'):
    print(name)
    print(group)
```

```python
df.groupby('Gender')['gross income']
```

```python
df.groupby('Gender')['gross income'].mean()
```

```python
#Hierarchical Grouping
df.groupby(['Gender','Customer type'])['gross income'].mean()
```

```python
#converting the above result to a DataFrame
df_new1 = df.groupby(['Gender','Customer type'])['gross income'].mean().reset_index()
df_new1.columns = ['Gender','Customer type','mean income']
df_new1
```

```python
def func(x):
    return x.max() - x.min()
```

```python
#you can also use user defined functions
df_new2 = df.groupby(['Gender','Customer type'])['gross income'].agg(func).reset_index()
df_new2.columns = ['Gender','Customer type','income difference']
df_new2
```

```python
#merging the dataframes
pd.merge(df_new1,df_new2,on=['Gender','Customer type'])
```

```python
#Another way to apply a function. This transforms the entire dataframe column directly
#It broadcasts the result to all the rows
df.groupby(['Gender','Customer type'])['gross income'].transform(np.mean)
```

```python
df['mean income'] = df.groupby(['Gender','Customer type'])['gross income'].transform(np.mean)
df
```

```python
#gives a similar result as the regular GroupBy flow
applied_data = df.head(20).groupby(['Gender','Customer type'])['gross income'].apply(lambda x: x.mean())
applied_data
```
