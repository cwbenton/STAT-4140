# Feedforward Neural Networks

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Generate simple dataset
X_train = np.random.rand(1000, 10)
y_train = np.random.rand(1000, 1)    
```

```python
# Define a simple feedforward neural network
# We have 10 input nodes, then three hidden layers of 64 nodes per layer, then a single output node.
model = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1)
])
```

```python
# Compile the model
# 'adam' is a commonly used optimizer. It helps achieve the global minimum at a moderate pace to find the minimum of the loss function.
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
'''
When doing neural networks, we don't send all the data at once. The common practice is to split the dataset
into small batches. This is where epoch=100 comes in. In this case, we will repeat the dataset 100 times.
'''
# Train the model
history = model.fit(X_train, y_train, epochs=100, validation_split=0.2, verbose=0)
```

```python
# Plot the training history
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
'''
This graph shows gross overfitting. We see our training loss goes down with more epochs, which is good; however, we can see that the validation
loss increases as the epochs go up. This is not good and is a classic case of overfitting in Neural Networks.
'''
```
